```python
#1. 함수를 제대로  호출하지 않음. n*2를 어디다 넣고 언제 뺄지도 정확하게 묘사 안됨
#2. 무엇을 return할지 명시 안함

def double(n):
    num = n * 2
    return num

double(5)
```

```python
sum = 0
while True:
    num = input()
    if num == '0':
        break
    sum += int(num)
print(f"총합은 {sum}이다")
```

```python
class Student:
    scores = []

    def add_score(self, score):
        self.scores.append(score)

s1 = Student()
s2 = Student()

s1.add_score(90)
s2.add_score(80)

print(s1.scores)
print(s2.scores)
```
