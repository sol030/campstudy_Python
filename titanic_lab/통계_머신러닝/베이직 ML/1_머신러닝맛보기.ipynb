# 머신러닝 맛보기 - 사과와 체리 분류

### 1-1. 단순 규칙기반 분류

```python
# 사과 지름에 따른 단순 분류 프로그램
fruit_diameter = 8.0

if fruit_diameter >= 7:         #지름이 7cm 이상이면 사과다
    print("사과")
```

### 1-2. 사과데이터 준비

```python
# 사과 데이터 (지름과 무게) / 머신러닝의 재료인 데이터를 리스트 형태로 준비 (지름과 무게라는 두가지 피처 사용)
apple_diameter = [7.2, 7.3, 7.5, 7.5, 7.6, 7.8, 7.8, 7.9,   ## 사과 지름 데이터
                  8.0, 8.0, 8.1, 8.2, 8.2, 8.2, 8.3, 8.3,
                  8.4, 8.4, 8.5, 8.5]
apple_weight = [140, 155, 170, 175, 185, 190, 200, 180,     ### 사과 무게 데이터
                195, 200, 195, 200, 205, 200, 170, 210,
                210, 220, 225, 215]
```

### 1-3. 산점도 그리기 (시각화)

```python
import matplotlib.pyplot as plt

# 사과 데이터 산점도
plt.scatter(apple_diameter, apple_weight)
plt.xlabel('diameter (cm)')
plt.ylabel('weight (g)')
plt.show()
```

### 1-4. 체리 데이터 준비


```python
# 체리 데이터 (지름과 무게)
cherry_diameter = [1.4, 1.5, 1.5, 1.6, 1.6, 1.7, 1.7, 1.7,
                   1.8, 1.8, 1.9, 1.9, 2.0, 2.1]
cherry_weight = [4, 5, 5, 6, 6, 7, 7, 6,
                 8, 7, 8, 9, 10, 11]
```

### 1-5. 사과 + 체리 산점도


```python
# 사과와 체리 데이터 함께 시각화
plt.scatter(apple_diameter, apple_weight)
plt.scatter(cherry_diameter, cherry_weight)
plt.xlabel('diameter (cm)')
plt.ylabel('weight (g)')
plt.show()
```

### 1-6. 데이터 합치기

```python
# 사과와 체리 데이터를 합치기
diameter = apple_diameter + cherry_diameter
weight = apple_weight + cherry_weight

# 2차원 리스트로 만들기
fruit_data = [[d, w] for d, w in zip(diameter, weight)]
print(fruit_data)
```

### 1-7. 정답(타깃) 데이터

```python
# 정답 데이터 (사과=1, 체리=0)
fruit_target = [1] * 20 + [0] * 14
print(fruit_target)
```

### 1-8. KNN 모델 훈련 & 평가


```python
from sklearn.neighbors import KNeighborsClassifier

# 모델 객체 생성 / KNeighborsClassifie로 이웃 데이터 보고 정체를 맞추는 KNN 모델의 객체를 만듦
kn = KNeighborsClassifier() #응? KNN이잖아 반갑다 얘 

# 데이터 훈련       / kn.fit(data, target)
kn.fit(fruit_data, fruit_target)    # training. 데이터와 정답을 모델에게 알려주며 학습시킴

# 정확도 평가       / kn.score() : 얼마나 잘 맞췄니? 
kn.score(fruit_data, fruit_target)  # → 1.0     # 1.0은 100% 다 맞혔다는 뜻
```
