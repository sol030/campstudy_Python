````python
# ì¤‘ì²© ë°˜ë³µë¬¸(Nested Loops)
for dan in range(2, 10):      # 2ë‹¨ë¶€í„° 9ë‹¨ê¹Œì§€
    for num in range(1, 10):  # ê° ë‹¨ì—ì„œ 1ë¶€í„° 9ê¹Œì§€ ê³±í•˜ê¸°
        print(f"{dan} x {num} = {dan * num}")
    print()  # ê° ë‹¨ì´ ëë‚˜ë©´ í•œ ì¤„ ê³µë°±
````

````python
#enumerate() ì‚¬ìš©í•´ë³´ê¸°
    #enumerate()ë¥¼ ì‚¬ìš©í•˜ë©´ ì¸ë±ìŠ¤ì™€ ê°’ì„ ë™ì‹œì— ì–»ì„ ìˆ˜ ìˆë‹¤
    #why? ë‹¨ìˆœ ê°’ì´ ì•„ë‹Œ ì¸ë±ìŠ¤ë„ í•¨ê»˜ ì‚¬ìš©í•˜ê³  ì‹¶ì„ ë•Œ ì‚¬ìš©!
fruits = "apple"

for idx, fruit_char in enumerate(fruits):
    print(f"{idx}: {fruit_char}")
````

````python
#zip() ì‚¬ìš©í•´ë³´ê¸°
    #zip()ì„ ì‚¬ìš©í•˜ë©´ ì—¬ëŸ¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ë³‘ë ¬ë¡œ ìˆœíšŒí•  ìˆ˜ ìˆë‹¤
    #why? ì—¬ëŸ¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ë™ì‹œì— ì‚¬ìš©í•˜ê³  ì‹¶ì„ ë•Œ ì‚¬ìš©!
fruits = ["apple", "banana", "cherry"]
prices = [1000, 2000, 3000]

for fruit, price in zip(fruits, prices):
    print(f"{fruit}: {price}ì›")
````
