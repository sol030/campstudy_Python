## 문제 1 — 인사와 글자수 분류

**사용 개념**: 입출력, 문자열, 조건문

이름을 입력받아 인사말과 이름의 글자수를 출력하고, 글자수가 5 이상이면 `이름이 길어요`, 아니면 `이름이 짧아요`를 출력하세요.

- **입력**: 한 줄 (이름)
- **출력**:
  1) `안녕하세요, 이름님!`
  2) `글자수: N`

## 문제 2 — 짝수/홀수 + 1부터 N까지 합

**사용 개념**: 입출력, 조건문, 반복문, 변수

정수 `N`을 입력받아 먼저 `짝수` 또는 `홀수`를 출력하고, 이어서 1부터 N까지의 **합**을 출력하세요.

- **입력**: 정수 N
- **출력**:
  1) `짝수` 또는 `홀수`
  2) `합: S`

**예시**
입력:
```
3
```
출력(예):
```
홀수
합: 6
```

````python
# TODO
N = int(input().strip())
sum = 0
# if로 짝/홀을 판단하고, for로 1부터 N까지의 합을 구해 출력하세요
if N % 2 == 0 :
    print('짝수')
else:
    print('홀수')
    
for i in range(1, N+1):   #1부터 n-1까지 / 왜? range에서도 n-1까지 셈
    sum += i
print(f"합: {sum}")
````

## 문제 10 — 로그인 (1회)

**사용 개념**: 입출력, 문자열, 조건문

아이디와 비밀번호를 한 번 입력받아 로그인 성공 여부를 출력하세요.

- **조건**:
  - 올바른 아이디: `admin` (대소문자 무시)
  - 올바른 비밀번호: `1234` (대소문자 구분)
- **출력**: 둘 다 맞으면 `성공`, 아니면 `실패`

**예시**
입력:
```
Admin
1234
```
출력:
```
성공
```

````python
# TODO
ID_OK = 'admin'
PW_OK = '1234'
# 입력을 받아 조건에 따라 '성공' 또는 '실패'를 출력하세요

id = input("아이디를 입력하시오").lower().strip()
pw = input("비밀번호를 입력하시오").strip()

if ID_OK == id and PW_OK == pw:
    print("들어가시오")
elif ID_OK != id or PW_OK != pw:
    print("나가시오")
else:
    print("무언가 잘못됐소")
````

````python
## 무한시도

ID_OK = 'admin'
PW_OK = '1234'
# 입력을 받아 조건에 따라 '성공' 또는 '실패'를 출력하세요

while True:
    id = input("아이디를 입력하시오").lower().strip()
    pw = input("비밀번호를 입력하시오").strip()
    if ID_OK == id and PW_OK == pw:
        print("들어가시오")
        break
    elif ID_OK != id or PW_OK != pw:
        print("나가시오")
    else:
        print("무언가 잘못됐소")
````
