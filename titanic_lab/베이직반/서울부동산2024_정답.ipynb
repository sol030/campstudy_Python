# 서울시 부동산 2024 (2024.csv) — 정답 및 해설

```python
import pandas as pd
try:
    df = pd.read_csv("2024.csv", encoding="cp949")
except UnicodeDecodeError:
    df = pd.read_csv("2024.csv", encoding="euc-kr")
print("컬럼:", df.columns.tolist())
df.head()
```

### 문제 1 정답 — 조건 필터링(AND)
- 해설: 두 조건을 괄호로 묶고 `&`로 결합합니다.

```python
sel1 = df[(df['자치구명'] == '강남구') & (df['건물용도'] == '오피스텔')]
print(sel1.head()); print("행 개수:", len(sel1))
```

### 문제 2 정답 — 조건 필터링(OR)
- 해설: 숫자 조건을 `|`로 결합합니다.

```python
sel2 = df[(df['층'] >= 10) | (df['물건금액(만원)'] >= 100000)]
print(sel2.head()); print("행 개수:", len(sel2))
```

### 문제 3 정답 — 결측치 조건
- 해설: `isnull()` 결과가 True인 행만 선택합니다.

```python
no_cancel = df[df['취소일'].isnull()]
print(no_cancel.head()); print("행 개수:", len(no_cancel))
```

### 문제 4 정답 — 특정 값 필터링
- 해설: 문자형 값 비교로 필터링합니다.

```python
direct = df[df['신고구분'] == '직거래']
print(direct.head()); print("행 개수:", len(direct))
```

### 문제 5 정답 — 정렬
- 해설: `sort_values`로 금액을 **내림차순** 정렬합니다.

```python
price_desc = df.sort_values(by='물건금액(만원)', ascending=False)
print(price_desc[['자치구명','법정동명','물건금액(만원)']].head(10))
```

### 문제 6 정답 — 다중 정렬
- 해설: `by`와 `ascending`에 각각 리스트를 전달해 다중 정렬합니다.

```python
multi_sorted = df.sort_values(by=['건축년도','건물면적(㎡)'], ascending=[True, False])
print(multi_sorted[['건축년도','건물면적(㎡)','자치구명','법정동명']].head(10))
```

### 문제 7 정답 — 열 제거
- 해설: `drop(columns=...)`로 새 DataFrame을 만듭니다.

```python
df_no_right = df.drop(columns=['권리구분'])
print("열 제거 후 컬럼:", df_no_right.columns.tolist())
```

### 문제 8 정답 — 열 이름 변경
- 해설: `rename(columns=...)`로 요구된 이름으로 변경합니다.

```python
df_ren = df.rename(columns={'물건금액(만원)':'거래금액_만원', '건물면적(㎡)':'건물면적_m2', '토지면적(㎡)':'토지면적_m2'})
print("변경 후 컬럼:", df_ren.columns.tolist())
```

### 문제 9 정답 — 결측치 개수
- 해설: `isnull().sum()`으로 각 열의 결측치 개수를 확인합니다.

```python
print(df.isnull().sum())
```

### 문제 10 정답 — 결측치 채우기(수치형)
- 해설: 중앙값으로 결측치를 채웁니다.

```python
median_year = df['건축년도'].median()
df_year_filled = df.copy()
df_year_filled['건축년도'] = df_year_filled['건축년도'].fillna(median_year)
print("채운 후 건축년도 결측치:", df_year_filled['건축년도'].isnull().sum())
```

### 문제 10-2 — 결측치 채우기(수치형)
- 해설: 자치구명별로 건축년도 중앙값으로 결측치를 채웁니다. 다만, 자치구 어떤 곳은 건축년도가 전부 결측이면 그 구의 중앙값도 NaN이라서 채우기가 안 될 수 있기에 그럴 땐 전체 중앙값으로 한 번 더 채우면 깔끔합니다!

```python
df_year_filled = df.copy()

gu_median = df_year_filled.groupby("자치구명")["건축년도"].transform("median")
global_median = df_year_filled["건축년도"].median()

df_year_filled["건축년도"] = (
    df_year_filled["건축년도"]
    .fillna(gu_median)
    .fillna(global_median)
)

print("채운 후 건축년도 결측치:", df_year_filled["건축년도"].isna().sum())
```

### 문제 11 정답 — 결측치 채우기(문자열)
- 해설: 문자형 컬럼은 적절한 텍스트로 채울 수 있습니다.

```python
df_name_filled = df.copy()
df_name_filled['건물명'] = df_name_filled['건물명'].fillna('미기재')
print("채운 후 건물명 결측치:", df_name_filled['건물명'].isnull().sum())
```

### 문제 12 정답 — 중복 확인 및 제거(subset)
- 해설: `subset`을 주어 중복 판정 기준을 지정하고 제거 전/후를 비교합니다.

```python
subset_cols = ['자치구명','법정동명','계약일','건물면적(㎡)','물건금액(만원)']
dup_cnt = df.duplicated(subset=subset_cols).sum()
df_unique = df.drop_duplicates(subset=subset_cols, keep='first')
print("중복 개수:", dup_cnt)
print("제거 전/후:", len(df), "->", len(df_unique))
```
