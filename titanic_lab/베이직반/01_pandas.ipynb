결측치(Missing Value)란 말 그대로 비어 있는 값 →NaN (Not a Number) 또는 NaT,  None으로 표시

결측치를 다루는 일반적인 방법은 두 가지:

1. **결측치가 있는 행/열 삭제** – 결측치가 극히 일부거나 해당 행/열이 크게 중요하지 않다면 아예 제거
2. **결측치를 특정 값으로 대체 (대치, imputation)** – 결측치가 많은 경우 삭제하면 데이터 손실 ⏫ 
    1. 수치형 변수 : 평균값이나 최빈값 등 **합리적인 값으로 채워 넣기**
    2. 범주형 변수 : “Unknown” 이라는 **합리적인 값으로 채워 넣기**

어떤 방법을 택할지는 상황에 따라 다릅니다. 

- 결측이 **적은 비율**이면 삭제가 간편하고, **많은 비율**이면 대체를 고려
- 결측치가 **어떻게 발생**했는지 (무작위인지 어떤 패턴인지) 도 중요한 판단 기준 → 현재는 모름

## 중복 제거 반드시 해야지

- 중복 데이터란 동일한 값의 행이 데이터셋에 반복되어 나타나는 경우
- 중복이 생기는 이유는 여러 가지인데
    - 데이터를 여러 소스로부터 합치는 과정에서 **같은 사람이 두 번 기록**되었을 때,
    - 센서 오작동 등으로 **동일한 로그가 반복** 저장되었을 때,
    - 크롤링이나 입력 실수로 **똑같은 행이 중복 입력**되었을 때 등이 있습니다.

**특정 케이스가 과대대표**되어 잘못된 결과를 초래 → 어떤 고객이 이중으로 기록되어 구매 횟수가 2배로 집계된다면 실제보다 부풀려진 통계 발생

````python
df3 = pd.DataFrame({'제품': ['A', 'B', 'B', 'C'],
                    '가격': [1000, 2000, 2000, 3000]})
print(df3)
````

````python
df3.duplicated(keep="first").sum()
#keep = first : 첫번째로 등장한 애만 보겠다 (중복은 안 보겠다)
#last : 마지막인 애들을 남긴다 (마지막만 보겟다)
#false : 중복행을 모두 제거하겠다 (오마에오 하이죠스루)
````

````python
df3_cleaned = df3.drop_duplicates() #s구나 헤헤
````

- **중복 여부 확인**
    - `df.duplicated()` 각 행이 이전에 이미 나타난 행과 중복인지 True/False로 표시하는 불린 Series를 반환
    - 기본적으로 **첫 등장 행은 False, 그 다음부터의 중복행은 True**로 표시
    - `df.duplicated().sum()` 을 사용하면 전체 중복 행 개수 파악

- **중복 행 제거**
    - `df.drop_duplicates()` **중복된 행을 제거**하고 **유일(unique)한 행만 남긴** DataFrame을 반환
    - 기본 설정에서는 **중복이 발생하면 가장 처음 나타난 행을 남기고 그 이후 중복된 행들을 제거**(`keep='first'`).

1. **완전 중복(행 전체가 동일)**
    
    → `drop_duplicates()`로 끝
    
2. **키 중복(예: PassengerId는 같지만 다른 컬럼이 다름)**
    
    → 어느 행을 남길지 규칙이 필요
    
    예:
    
    - 결측이 더 적은 행을 남긴다
    - 최신 타임스탬프를 남긴다
    - 특정 컬럼을 우선한다

**예: PassengerId가 같은 행 중 “결측이 더 적은” 행을 남기기**
시간이 필요하니까 최근 걸 남기겠다
Fare니까 높은 값을 남겨야겠다

이런 조건을 본인이 세워야함

````python
temp = dup_df.copy()
````

````python
temp.isna().sum()
````

````python
temp.isna().sum(axis=1) #axis=1로 행기준으로 볼 수 있음(행단위로)
````

````python
temp['na_count'] = temp.isna().sum(axis=1)
temp
````

````python
## 결측 적은 애를 남길 거니까 일단 정렬

temp.sort_values(by=['PassengerId', 'na_count'])    #as어쩌구 기본 : TRue
````

````python
temp.sort_values(by=['PassengerId', 'na_count']).drop_duplicates(subset=['PassengerId'])        

#정렬을 PassengerId(1순위), na_count(2순위)로 할 거야
#subset로 'PassengerId' 기준으로 드랍할 건데~ 라는 거

##drop_duplicates(subset=['PassengerId']) : 이제 같은 승객 번호를 가진 데이터들 중 맨 위에 있는 것만 남기고 나머지는 버리는 거
## 숨어있는 keep=first : 판다스의 기본값 (keep='first'): 중복된 것들 중 가장 첫 번째(맨 위) 행만 남김
````
