### 문제 10 — 결측치 채우기(수치형)
- 건축년도 결측치를 건축년도 **중앙값**으로 채운 새 DataFrame을 만들고, 채운 후 해당 열의 결측치 개수를 출력하세요.

```python
# TODO: 여기에서 코드를 작성하세요.
# df2 = df2['건축년도'].fillna(df2['건축년도'].median())
df2['건축년도'] = df2['건축년도'].fillna(df2['건축년도'].median())
#아래처럼 해도 동일함
df2['건축년도'].fillna(df2['건축년도'].median(), inplace=True)  #재적용 안해도 바로 되도록
df2['건축년도'].isna().sum()
```

```python
df3 = df.copy()
```

```python
# TODO: 여기에서 코드를 작성하세요.
# df3['건축년도'] = df3['건축년도'].fillna(df['자치구명'].median())
#결측치를 자치구명'별로' 건축년도 '중앙값으로' 채우다 그럼 fillna는 건축년도고, 어딘가에 지금 자치구명이 들어가야하는데?
#~~별로는 groupby아니면 컬럼 다음에 [] 넣는 거 아닌가? 
# 아 기억났다!!!!! 걍 두개넣는거잖아저거
# 여성남성 별로 결측치 채우기 이거 했었는데 아까

df3['건축년도'] = df3['건축년도'].fillna(df['건축년도'].median())
df3['건축년도'].isna().sum()
```

### 문제 11 — 결측치 채우기(문자열)

```python
# TODO: 여기에서 코드를 작성하세요.
### 줌 강의 내용 다시 들어야 함 이거
df3['건물명'] = df3['건물명'].fillna('미기재')
df3['건물명'].isna().sum()
```

### 문제 12 — 중복 확인 및 제거(subset)
