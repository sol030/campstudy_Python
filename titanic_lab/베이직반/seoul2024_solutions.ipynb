# [정답/해설] 서울시 부동산 2024 — 기본 전처리 & 인덱싱/슬라이싱

```python
import pandas as pd
path = "2024.csv"
df = pd.read_csv(path, encoding="cp949")
print("로드 완료! shape:", df.shape)
```

### 1. `head()` — 데이터 미리보기

```python
df.head()
```

**해설:** 구조/열 이름을 빠르게 확인합니다.

### 2. `shape` — (행, 열) 크기

```python
df.shape
```

**해설:** 행/열 수를 파악합니다.

### 3. `info()` — 타입 & 결측 개요

```python
df.info()
```

**해설:** 각 컬럼의 dtype/결측 개수를 한 번에 확인합니다.

### 4. `describe()` — 수치형 요약 통계

```python
df.describe()
```

**해설:** 범위/분포/이상치 감을 잡습니다.

### 5. 열 선택 — 리스트로 여러 열

```python
df[['자치구명','법정동명','계약일','물건금액(만원)','건물면적(㎡)']].head()
```

**해설:** `[['col1','col2']]`는 **DataFrame**을 반환합니다.

### 6. `to_datetime`으로 날짜형 변환 + `info()` 재확인

```python
df['계약일'] = pd.to_datetime(df['계약일'])
df.info()
```

**해설:** 날짜형으로 변환하면 시계열 연산이 가능해집니다.

### 7. `.loc` — 라벨 기반 슬라이싱(끝 포함)

```python
df.loc[0:9, ['자치구명','법정동명','계약일','물건금액(만원)','건물면적(㎡)']]
```

**해설:** `.loc[a:b]`는 **b를 포함**합니다.

### 8. `.iloc` — 정수 위치 기반 슬라이싱(끝 제외)

```python
df.iloc[0:10, [0,1,6,7,8]]
```

**해설:** `.iloc[a:b]`는 **b를 제외**합니다.

### 9. 불리언 필터 — 강남구 & 85㎡ 이상

```python
mask = (df['자치구명'] == '강남구') & (df['건물면적(㎡)'] >= 85)
df.loc[mask, ['자치구명','법정동명','건물면적(㎡)','물건금액(만원)']].head()
```

**해설:** 조건이 True인 행만 남깁니다.

### 10. 불리언 필터 — 건축년도 2000년 이후

```python
df.loc[df['건축년도'] >= 2000, ['자치구명','건축년도','건물용도']].head()
```

**해설:** 최신 건축물만 추립니다.

### 11. `value_counts()` — 범주 분포 (상위 5)

```python
df['건물용도'].value_counts().head()
```

**해설:** 주요 용도의 빈도를 파악합니다.

### 12. `isnull().sum()` — 결측 개수

```python
df.isnull().sum()
```

**해설:** 결측치 처리 전략 수립에 필요합니다.

### 13. `to_csv` — 필터 결과 저장

```python
mask = (df['자치구명'] == '강남구') & (df['건물면적(㎡)'] >= 85)
out = df.loc[mask, ['자치구명','법정동명','건물면적(㎡)','물건금액(만원)']]
out.to_csv('seoul2024_gangnam_85.csv', index=False)
print('saved: seoul2024_gangnam_85.csv')
```

**해설:** 전처리 산출물을 파일로 기록합니다.
