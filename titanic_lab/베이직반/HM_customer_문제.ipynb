### 문제 2 — 다중 조건
- 클럽 회원 상태가 'ACTIVE' 이면서 `FN` 값이 1인 고객만 필터링하세요.

### 문제 3 — 다중 조건
- 나이가 18세 미만 또는 65세 이상인 고객만 필터링하세요.

```python
# TODO: 여기에서 코드를 작성하세요.
# 뭐더라 그 isin하는거 있는데 아까했는데 다중조건 연결하는거 보고와야지 안되네 뭐 안에 뭐 있는지 보는거지 참
df[(df['age'] < 18) | (df['age'] >= 65)]
```
