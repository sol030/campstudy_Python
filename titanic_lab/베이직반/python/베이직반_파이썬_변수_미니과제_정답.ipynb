## 문제 1
다음 코드의 출력 결과를 예측해보세요.

````python
x = 5
y = x
x = 10
print(y)
````

해설
- 처음에 x에 5가 할당됩니다.
- y에 x의 값인 5가 할당됩니다.
- 이후 x에 10이 재할당되지만, y는 여전히 5를 가지고 있습니다.
- 따라서 print(y)는 5를 출력합니다.

deep
- 정수형(int)은 불변(Immutable) 객체입니다. 즉, 값이 변경되지 않습니다.
- 변수에 새로운 값을 할당하면 변수는 새로운 객체를 가리키게 됩니다.
- 변수 x에 10을 할당하면 x는 이제 10 객체를 가리키지만, y는 여전히 5를 가리킵니다.

### 문제 1번 DeepDive
y가 항상 최신화된 x의 값을 가지고 싶다면?  

1. 변경 가능한(Mutable) 객체 사용하기 -> 리스트나 딕셔너리
```python
x = [5]  # 리스트로 감싸서 변경 가능한 객체로 만듭니다.
y = x    # y와 x는 같은 리스트 객체를 가리킵니다.

x[0] = 10  # 리스트의 첫 번째 요소를 변경합니다.

print(y[0])  # 출력: 10
```

2. 클래스를 사용하여 속성 공유하기
```python
class Number:
    def __init__(self, value):
        self.value = value

x = Number(5)
y = x  # x와 y는 동일한 Number 객체를 가리킵니다.

x.value = 10  # 객체의 속성 value를 변경합니다.

print(y.value)  # 출력: 10
```

3. 함수를 사용하여 동적으로 값 가져오기
```python
x = 5

def y():
    return x

x = 10
print(y())  # 출력: 10
```

## 문제 7
파이썬에서 변수 PI를 선언하여 원주율 3.14159를 저장하고, 반지름이 5인 원의 면적을 계산하여 출력하는 코드를 작성하세요. (원의 면적 = PI * r * r)

````python
PI = 3.14159
radius = 5
area = PI * radius * radius
print(f"반지름이 {radius}인 원의 면적은 {area}입니다.")
````

해설
- PI 변수에 원주율 3.14159를 저장합니다.
- radius 변수에 반지름 5를 저장합니다.
- 원의 면적은 PI * 반지름 * 반지름으로 계산합니다.
- print() 함수를 사용하여 결과를 출력합니다.

## 문제 9
변수 x에 숫자 10을 저장하고, 변수 y에 문자열 '10'을 저장했습니다.  
이 두 변수가 같은 값인지, 같은 데이터 타입인지 비교하는 코드를 작성하고 결과를 출력하세요.

````python
x = 10
y = '10'

print(f"x와 y의 값이 같은 값인가요? : {x==y}")
print(f"x와 y의 값이 같은 데이터 타입인가요? : {type(x)==type(y)}")
````

해설
- x는 정수형 int이고, y는 문자열형 str입니다.
- 값 비교 x == y는 10 == '10'으로 서로 다른 타입이므로 False입니다.
- 데이터 타입 비교 type(x) == type(y)는 int == str이므로 False입니다.
