##### 전처리 체크리스트

3. **결측치 패턴은?** (어디에, 얼마나, 어떤 조건에서 비는지)
4. **이상치/오류는?** (나이 -1, 요금 999999, 문자열 섞임)

8. **인덱스는 의미가 있나?** (0..n-1 유지 vs 키로 설정)
