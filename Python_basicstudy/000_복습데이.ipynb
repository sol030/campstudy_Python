````python
d1 = {"x": [1, 2], "y": [3, 4]}
d2 = d1.copy()
d1["x"].append(999)
print(d1)
print(d2)

d3 = {"x": (1, 2), "y": (3, 4)}
d4 = d3.copy()
# d3["x"] += (999,)
````

````python
# 가변 객체(리스트)를 값으로 가진 딕셔너리
d1 = {"x": [1, 2], "y": [3, 4]}
d2 = d1.copy()          # 얕은 복사 → 딕셔너리 객체만 새로 생성, 값은 같은 리스트 참조

d1["x"].append(999)     # d1["x"] 리스트 수정 → d2["x"]에도 반영됨
print(d1)               # {'x': [1, 2, 999], 'y': [3, 4]}
print(d2)               # {'x': [1, 2, 999], 'y': [3, 4]}
````

````python
# 불변 객체(튜플)를 값으로 가진 딕셔너리
d3 = {"x": (1, 2), "y": (3, 4)}
d4 = d3.copy()          # 얕은 복사 → 딕셔너리 객체만 새로 생성

d3["x"] += (999,)       # 튜플은 불변 → 새 튜플 생성, d4["x"]는 원래 튜플 유지
print(d3)               # {'x': (1, 2, 999), 'y': (3, 4)}
print(d4)               # {'x': (1, 2), 'y': (3, 4)}
````

````python

# 깊은 복사 예시
import copy
d5 = copy.deepcopy(d1)  # 리스트까지 새 객체 생성
d1["x"].append(1000)
print(d1)               # {'x': [1, 2, 999, 1000], 'y': [3, 4]}
print(d5)               # {'x': [1, 2, 999], 'y': [3, 4]} - 독립적
````

````python
#2
price_table = {
    "A100": 12000,
    "B200": 3500,
    "C300": 19900,
    "D400": 55000,
    "E500": 9900,
}

coupon_rules = {
    "A100": 10,   # 10% off
    "C300": 15,
    "E500": 5,
}

shipping_policy = {
    "fee": 3000,
    "free_over": 50000,
}

items = [("A100", 2), ("B200", 5), ("C300", 1), ("E500", 3)]

invoice = {"lines": []}  # 항목별 상세 내역
subtotal_after_discount = 0  # 할인 적용 후 합계 누적

for code, qty in items:
    # 1. 상품 단가 가져오기
    unit_price = price_table[code]

    # 2. line_subtotal 계산 (단가 * 수량)
    line_subtotal = unit_price * qty

    # 3. 상품별 할인율 가져오기, 없으면 0
    discount_rate = coupon_rules.get(code, 0)

    # 4. line_total 계산 (정률 할인 적용)
    # 할인율 % 적용 → (100 - discount_rate)/100 곱함
    line_total = int(line_subtotal * (100 - discount_rate) / 100)

    # 5. invoice["lines"]에 튜플 추가
    invoice["lines"].append(
        (code, qty, unit_price, line_subtotal, discount_rate, line_total)
    )

    # 6. subtotal_after_discount 누적
    subtotal_after_discount += line_total

# 7. 배송비 계산
shipping_fee = 0 if subtotal_after_discount >= shipping_policy["free_over"] else shipping_policy["fee"]

# 8. grand_total 계산 및 invoice에 저장
invoice["subtotal_after_discount"] = subtotal_after_discount
invoice["shipping_fee"] = shipping_fee
invoice["grand_total"] = subtotal_after_discount + shipping_fee

# 출력
print(invoice)
````

````python
sessions = [
    ("s01", "u01", 1700),
    ("s02", "u01", 1750),
    ("s03", "u02", 1600),
    ("s04", "u03", 1900),
    ("s05", "u02", 1650),
    ("s06", "u04", 1200),
    ("s07", "u04", 2100),
    ("s08", "u05", 1300),
    ("s09", "u05", 1299),
    ("s10", "u06", 2200),
    ("s11", "u06", 2199),
]

expire_before = 1500

active_sessions = {}
expired_sessions = []

for session_id, user_id, last_seen in sessions:
    # TODO 1: 만료 세션이면 expired_sessions에 추가하고 다음으로
    if last_seen < expire_before:
        expired_sessions.append(session_id)
        continue
    
    # TODO 2: 만료가 아니면 active_sessions에서 user_id 기준으로 최신 세션만 유지
    if user_id not in active_sessions:
        active_sessions[user_id] = (session_id, last_seen)
    else:
        # 기존 세션과 비교하여 더 최신이면 교체
        if last_seen > active_sessions[user_id][1]:
            active_sessions[user_id] = (session_id, last_seen)

print(active_sessions)
print(sorted(expired_sessions))
````

````python
# 학생별 성적 데이터를 딕셔너리로 정의
# key: 학생 이름, value: 점수 리스트([국어, 영어, 수학])
students = {
    "민수": [80, 70, 90],
    "영희": [40, 55, 60],
    "지수": [100, 95, 90],
    "철수": [30, 45, 20]
}

# 점수 리스트를 받아 평균을 계산하는 함수
def calc_avg(scores):
    # sum(scores)로 총점 계산, len(scores)로 과목 수 나누어 평균 계산
    return sum(scores) / len(scores)

# 학생별 평균 점수 딕셔너리 생성
def make_avg_dict(students):
    result = {}  # 결과를 저장할 빈 딕셔너리 생성
    for name in students:  # 각 학생 이름 순회
        # calc_avg 함수를 이용해 각 학생 점수 리스트의 평균 계산 후 딕셔너리에 저장
        result[name] = calc_avg(students[name])
    return result  # 학생별 평균 딕셔너리 반환

# 평균이 60점 이상인 학생만 추출
def get_pass(avg_dict):
    result = {}  # 합격자 결과 저장용 딕셔너리
    for name in avg_dict:  # 학생별 평균 점수 순회
        if avg_dict[name] >= 60:  # 평균이 60점 이상이면
            result[name] = avg_dict[name]  # 합격자 딕셔너리에 추가
    return result  # 합격자 딕셔너리 반환

# 최고 평균 학생 찾기
def get_top_student(avg_dict):
    max_name = ""   # 최고 점수 학생 이름 초기화
    max_score = 0   # 최고 점수 초기화
    for name in avg_dict:  # 학생별 평균 점수 순회
        if avg_dict[name] > max_score:  # 현재 학생 점수가 최고 점수보다 크면
            max_score = avg_dict[name]  # 최고 점수 갱신
            max_name = name            # 최고 점수 학생 이름 갱신
    return max_name  # 최고 평균 학생 이름 반환

# 최고 평균 학생 찾기
##다른 방식(내장함수 활용)
def get_top_student(avg_dict):
    return max(avg_dict, key=avg_dict.get)

# 학생별 평균 점수 계산
avg_dict = make_avg_dict(students)
print("학생별 평균:", avg_dict)  # 학생별 평균 출력

# 평균 60점 이상 합격자 출력
print("합격자:", get_pass(avg_dict))

# 최고 평균 학생 출력
print("최고 평균 학생:", get_top_student(avg_dict))
````

````python
# 입금 함수
def deposit(accounts, acc, money):
    # accounts 딕셔너리에서 acc 계좌에 money 만큼 더함
    accounts[acc] += money

# 출금 함수
def withdraw(accounts, acc, money):
    # 잔액이 부족하면 출금 실패 메시지 출력 후 종료
    if accounts[acc] < money:
        print(acc, "출금 실패")
        return
    # 충분하면 잔액에서 money 차감
    accounts[acc] -= money

# 은행 전체 자산 계산
def total_money(accounts):
    # 딕셔너리 값(잔액)들을 모두 더함
    return sum(accounts.values())

# 위험 계좌 확인 함수
def risky_accounts(accounts):
    result = []
    for acc in accounts:
        # 잔액이 5만원 미만이면 위험 계좌로 간주
        if accounts[acc] < 50000:
            result.append(acc)
    return result

# 초기 계좌 상태
accounts = {
    "A001": 50000,
    "A002": 120000,
    "A003": 30000
}

# A001 계좌에 2만원 입금
deposit(accounts, "A001", 20000)
print("A001 잔액:", accounts["A001"])  # 입금 후 잔액 확인

# A003 계좌에서 5만원 출금 시도 (잔액 부족)
withdraw(accounts, "A003", 50000)  # 실패 메시지 출력

# 전체 은행 자산 출력
print("전체 은행 자산:", total_money(accounts))

# 위험 계좌(5만원 미만) 출력
print("위험 계좌:", risky_accounts(accounts))
````

````python
# 문자열 로그를 단어 리스트로 변환
def make_list(log):
    # 공백 기준으로 문자열을 나눠 리스트 반환
    return log.split()

# 단어별 등장 횟수 계산
def make_count(words):
    result = {}
    for w in words:
        if w in result:
            # 이미 존재하면 1 증가
            result[w] += 1
        else:
            # 처음 등장하면 1로 초기화
            result[w] = 1
    return result

# 가장 많이 등장한 단어 반환
def most_word(counts):
    # 딕셔너리에서 value 기준 최대 key 반환
    return max(counts, key=counts.get)

# 특정 단어의 전체 대비 비율 계산
def ratio(counts, word):
    total = sum(counts.values())  # 전체 단어 수
    return counts[word] / total * 100  # 비율 계산

# 중요 단어 추출 (출현 횟수 2회 이상)
def important(counts):
    result = {}
    for k in counts:
        if counts[k] >= 2:
            result[k] = counts[k]
    return result

# 분석할 로그 문자열
log = "error login success login error logout success login error"

# 로그를 단어 리스트로 변환
words = make_list(log)

# 단어별 등장 횟수 계산
counts = make_count(words)

# 결과 출력
print("단어 빈도:", counts)  # {'error': 3, 'login': 3, 'success': 2, 'logout': 1}
print("최다 단어:", most_word(counts))  # 'error' 또는 'login' (같은 최대값이면 먼저 등장한 key)
print("error 비율:", round(ratio(counts, "error"), 1), "%")  # 3/9 * 100 = 33.3%
print("중요 단어:", important(counts))  # {'error': 3, 'login': 3, 'success': 2}
````
