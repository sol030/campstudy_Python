```python
height_cm = int(input("키를 입력하세요."))
hight_m = height_cm / 100
weight_kg = int(input("몸무게를 입력하세요."))
bmi = weight_kg / (hight_m)**2
if bmi >= 0 and bmi < 18.5:
    print(f"BMI: {bmi} → 저체중")
elif bmi >= 18.5 and bmi < 23:
    print(f"BMI: {bmi} → 정상")
elif bmi >= 23 and bmi < 25:
    print(f"BMI: {bmi} → 과체중")
elif bmi >= 25:
    print(f"BMI: {bmi} → 비만")
else:
    print(f"잘못된 입력입니다.") 
```

```python
a, b, c = map(int, input().split())  #리스트로 이것저것 할거니까 list(map()) 해도 되나? 했는데
if a<=0 or b<=0 or c<=0 or (a+b <= c or a+c <= b or b+c <= a): #굳이 list로 안 묶어도 됨 map() 순회 가능한 객체라 굳이굳이
    print("삼각형이 아닙니다.")
elif (a**2 + b**2) == c**2:
    print("직각삼각형")
elif a == b == c:
    print("정삼각형")
elif a == b or b == c:
    print("이등변삼각형")
else:
    print("일반삼각형")
```
