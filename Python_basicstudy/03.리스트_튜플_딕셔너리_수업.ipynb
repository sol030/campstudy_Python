````python
scores = {"Alice": 88, "Bob": 95, "Charlie": 70, "Dave": 95, "Eve": 62}

avg = sum(scores.values()) / len(scores)
print(f"í‰ê·  ì ìˆ˜: {avg}")

max_score = max(scores.values())
top_students = [f"{name}({score})" for name, score in scores.items() if score == max_score]
print("ìµœê³  ì ìˆ˜ í•™ìƒ:", ", ".join(top_students))

result_for = {}
for name, score in scores.items():
    if score >= 70:
        result_for[name] = "Pass"
    else:
        result_for[name] = "Fail"

print("ì¼ë°˜ forë¬¸:", result_for)

result_comp = {name: "Pass" if score >= 70 else "Fail" for name, score in scores.items()}
print("ì»´í”„ë¦¬í—¨ì…˜:", result_comp)
````
