### 1-3) `end` 옵션: 출력의 “끝”에 붙는 것

## 2) 문자열 채우기: `format()` vs f-string

출력할 때 이런 문장을 자주 만들고 싶습니다.

- “적토마의 점수는 95점입니다.”
- “가격은 1,234,567원입니다.”
- “평균은 3.14입니다.”

즉, **문장 안에 변수 값을 끼워 넣기**가 필요합니다.

---

### 2-1) `format()` 방식
문자열 안에 `{}`를 만들고, 뒤에서 `.format(...)`로 채웁니다.


#### ✅ `format()` 예제 (4개)

```python
name = "적토마"
score = 95
print("{}의 점수는 {}점입니다.".format(name, score))
```

```python
pi = 3.141592
print("원주율은 {:.2f}입니다.".format(pi))  # 소수 둘째 자리까지
```

```python
money = 1234567
print("가격은 {:,}원입니다.".format(money))  # 3자리마다 콤마
```

```python
# 같은 값을 여러 번 쓰고 싶을 때
item = "연필"
print("저는 {}을(를) 좋아합니다. {}은(는) 유용합니다.".format(item, item))
```

### 2-2) f-string 방식 (요즘 가장 많이 사용)
문자열 앞에 `f`를 붙이고, `{변수}`처럼 바로 넣습니다.

- 코드가 더 읽기 쉽습니다.
- 작성이 빠릅니다.


#### ✅ f-string 예제 (5개)

```python
name = "적토마"
score = 95
print(f"{name}의 점수는 {score}점입니다.")
```

```python
money = 1234567
print(f"가격은 {money:,}원입니다.")
```

```python
pi = 3.141592
print(f"원주율은 {pi:.2f}입니다.")
```

```python
# 계산 결과를 바로 넣을 수도 있습니다.
a, b = 7, 3
print(f"{a} + {b} = {a + b}")
```

```python
# 소수 둘째 자리 + 콤마를 같이 쓸 수도 있습니다.
big = 1234567.891
print(f"큰 수: {big:,.2f}")
```

### 2-3) 헷갈리기 쉬운 상황(중요)

#### (1) f-string은 앞에 `f`를 꼭 붙여야 합니다
붙이지 않으면 `{name}`이 그냥 글자 그대로 출력됩니다.


```python
name = "적토마"
print("{name}님, 환영합니다!")   # f가 없으면 그대로 출력됩니다.
print(f"{name}님, 환영합니다!")  # f가 있으면 값으로 바뀝니다.
```

#### (2) 중괄호 `{}`는 특별한 의미가 있습니다
문장에 중괄호를 진짜로 출력하고 싶으면 `{{` `}}`처럼 두 번 써야 합니다(심화지만 자주 마주칩니다).


```python
x = 10
print(f"x는 {x}입니다. {{이건 중괄호}}")
```

#### (2) 사용자가 숫자가 아닌 것을 입력하면?
예: `hello`, `3.14`를 int로 바꾸기, `1,000`처럼 콤마 포함 등

이때 `int()`/`float()`는 **에러**를 냅니다.  
프로그램이 멈추지 않게 하려면 `try/except`를 사용합니다.


#### ✅ `try/except` 예제 (5개)

```python
# 예제 1: 정수 변환 실패를 잡기
s = input("정수를 입력하세요: ")
try:
    n = int(s)
    print("n * 2 =", n * 2)
except ValueError:
    print("숫자가 아니라서 변환에 실패했습니다!")
```

```python
# 예제 2: 실수 변환 실패를 잡기
s = input("실수를 입력하세요(예: 3.14): ")
try:
    x = float(s)
    print("x + 1 =", x + 1)
except ValueError:
    print("실수로 바꿀 수 없는 입력입니다!")
```

```python
# 예제 3: 0으로 나누기 에러도 잡기
s = input("정수 2개를 공백으로 입력하세요(예: 10 0): ")
try:
    a, b = map(int, s.split())
    print("a / b =", a / b)
except ValueError:
    print("입력 형식이 잘못되었습니다! 예: 10 3")
except ZeroDivisionError:
    print("0으로는 나눌 수 없습니다!")
```

```python
# 예제 4: 콤마가 포함된 숫자 처리(간단 버전)
s = input("콤마 포함 숫자를 입력하세요(예: 1,234): ")
s2 = s.replace(",", "")  # 콤마 제거
try:
    n = int(s2)
    print("변환 성공:", n)
except ValueError:
    print("숫자로 바꿀 수 없습니다!")
```

```python
# 예제 5: 빈 입력 처리(간단)
s = input("아무것도 안 치고 엔터도 눌러보세요: ")
if s == "":
    print("빈 입력입니다. 다시 입력이 필요합니다.")
else:
    print("입력한 값:", s)
```

## 4) 한 줄에서 여러 값 입력받기: `split()` + `map()`

### 4-1) `split()`이란?
문자열을 **쪼개서** 여러 조각으로 만드는 기능입니다.

예를 들어:
- `"10 20 30".split()` → `["10", "20", "30"]` (공백 기준)
- `"a,b,c".split(",")` → `["a", "b", "c"]` (쉼표 기준)

#### ✅ `split()` 예제 (6개)


```python
print("10 20 30".split())
```

```python
print("a,b,c".split(","))
```

```python
print("Read#Analyze#Implement".split("#"))
```

```python
# 연속 공백은 split()가 알아서 정리해 줍니다.
print("10   20    30".split())
```

```python
# 하지만 split(',')은 공백을 자동으로 없애주지 않습니다.
print("10, 20, 30".split(","))
```

```python
# 공백이 거슬리면 strip()으로 정리할 수 있습니다.
parts = "10, 20, 30".split(",")
clean = [p.strip() for p in parts]
print(clean)
```

### 4-2) `map(int, ...)`란?
쪼갠 결과는 보통 문자열입니다. 예: `"10"`  
`map(int, ...)`는 각 조각에 `int()`를 **한 번에 적용**합니다.

자주 쓰는 형태:
- `a, b = map(int, input().split())`
- `nums = list(map(int, input().split(',')))`

#### ✅ `map(int, ...)` 예제 (6개)


```python
a, b = map(int, "3 7".split())
print("a+b =", a + b)
```

```python
nums = list(map(int, "1 2 3 4".split()))
print(nums, "합:", sum(nums))
```

```python
nums = list(map(int, "1,2,3".split(",")))
print(nums)
```

```python
# 쉼표 뒤 공백이 있는 경우: 먼저 strip 처리하기
parts = "10, 20, 30".split(",")
nums = list(map(int, [p.strip() for p in parts]))
print(nums)
```

```python
# 사용자 입력 예시(직접 입력): 10 20
a, b = map(int, input("정수 2개(공백): ").split())
print("곱:", a * b)
```

```python
# 사용자 입력 예시(직접 입력): 10#20#30
scores = list(map(int, input("정수 3개(#): ").split("#")))
print("최댓값:", max(scores), "최솟값:", min(scores))
```

### 4-3) 헷갈리기 쉬운 상황(매우 중요)

#### (1) split(',') 결과에 공백이 남을 수 있습니다
`"10, 20".split(",")` → `["10", " 20"]`  
이건 `int(" 20")`이 가능해서 보통 괜찮지만, 상황에 따라 공백이 문제를 만들기도 합니다.  
그래서 **습관적으로 `strip()`**을 하는 것이 안전합니다.

#### (2) 입력 개수가 다르면 에러가 납니다
`a, b = ...`는 **딱 2개**가 들어와야 합니다.  
1개 또는 3개가 들어오면 에러가 납니다.

#### (3) 구분자가 달라지면 split도 달라져야 합니다
- 공백 입력이면 `split()`
- 쉼표 입력이면 `split(',')`
- `#` 입력이면 `split('#')`


# 미니 프로젝트: “안전한 계산기” 만들기

### 목표
사용자가 두 수를 입력하면 아래를 출력합니다.

- 합, 차, 곱, 나눗셈(실수)
- 입력이 이상하면 친절한 안내를 출력하고 끝내기
- 0으로 나누면 멈추지 않게 처리하기

### 입력 형식
아래 중 어떤 형식이든 받도록 해 봅시다(가능하면):
- `10 3` (공백)
- `10,3` (쉼표)
- `10#3` (#)

힌트:
1) 입력 문자열에 `,`가 있으면 `split(',')`  
2) 아니면 `#`가 있으면 `split('#')`  
3) 아니면 `split()` (공백)

아래 코드를 완성해 보세요.


```python
s = input("두 수를 입력하세요 (예: 10 3 / 10,3 / 10#3): ")

# TODO: 구분자 선택
# TODO: 숫자 변환(필요하면 strip, replace 등)
# TODO: 합/차/곱/나눗셈 출력
# TODO: ValueError, ZeroDivisionError 처리

try:
    if ',' in s:
        parts = s.split(',')
    elif '#' in s:
        parts = s.split('#')
    else:
        parts = s.split()

    parts = [p.strip() for p in parts]
    a, b = map(float, parts)

    print(f"합: {a + b}")
    print(f"차: {a - b}")
    print(f"곱: {a * b}")
    print(f"나눗셈: {a / b}")

except ValueError:
    print("두 개의 숫자를 올바르게 입력해주세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
```

```python
# 예시 정답(일부)

# sep/end 예시
print("A", "B", "C", sep="|", end="...")
print("DONE")

# f-string 서식 예시
money = 9876543
print(f"총액: {money:,}원")

# input + 형변환 예시
a = int(input("정수 a: "))
b = int(input("정수 b: "))
print(a + b)

# split/map + strip 예시
s = "10, 20, 30"
parts = [p.strip() for p in s.split(",")]
nums = list(map(int, parts))
print(nums, sum(nums))
```

## 오늘 정리 체크리스트

- [ ] `print(sep=..., end=...)`로 출력 모양을 바꿀 수 있습니다.  
- [ ] f-string과 format으로 문자열에 값을 채워 넣을 수 있습니다.  
- [ ] `input()`은 **항상 문자열**이므로 숫자로 쓰려면 변환이 필요합니다.  
- [ ] `split()`과 `map(int, ...)`로 한 줄에서 여러 숫자를 읽을 수 있습니다.  
- [ ] `try/except`로 입력이 이상해도 프로그램이 멈추지 않게 만들 수 있습니다.
