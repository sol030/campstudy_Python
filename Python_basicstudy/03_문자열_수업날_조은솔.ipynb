## 1) 문자열은 불변(immutable)입니다

“불변(immutable)”은 **한 번 만든 문자열은 ‘그 자리에서’ 수정할 수 없다**는 뜻입니다.

예를 들어, `"hello"`에서 첫 글자 `h`를 `H`로 바꾸고 싶다고 해 봅시다.  
많은 초보자분들이 아래처럼 시도합니다:

```python
word = "hello"
word[0] = "H"   # ❌ 에러
```

왜 에러일까요?
- 문자열은 “글자들을 담은 상자”처럼 보이지만
- 파이썬에서는 **수정 불가능한 타입**으로 설계되어 있습니다.

대신, **새 문자열을 만들어서 다시 저장(재할당)** 해야 합니다.


````python
word = "hello"
print(word)

# 아래는 에러 예시입니다(주석을 풀면 TypeError).
# word[0] = "H"

print("에러 예시는 주석으로 막아두었습니다.")
````

### 1-1) 문자열을 “수정”하고 싶을 때의 정답: 새 문자열 만들기

방법 예시:
- `replace()` 사용
- 슬라이싱으로 조합해서 새 문자열 만들기

````python
# 예제 3: 끝에 느낌표 붙이기
msg = "good"
msg2 = msg + "!"
print(msg, "->", msg2)
````

### 1-2) 헷갈리기 쉬운 상황(중요)

`text.strip()`을 했는데, 왜 `text`가 그대로일 때가 있을까요?  
답: `strip()`은 **새 문자열을 반환**합니다.  
그래서 `text = text.strip()`처럼 **다시 저장**해야 합니다.


````python
text = "   hello   "
text.strip()          # 결과를 버림(저장 안 함)
print("저장 안 함:", repr(text))

text = text.strip()   # 결과를 다시 저장
print("저장 함:", repr(text))
````

## 2) 인덱싱(Indexing)과 슬라이싱(Slicing)

문자열의 각 글자에는 **번호(인덱스)** 가 있습니다.

예: `s = "Python"`

- `s[0]` → `"P"` (첫 글자)
- `s[1]` → `"y"`
- `s[5]` → `"n"` (마지막 글자)

그리고 뒤에서부터도 셀 수 있습니다:
- `s[-1]` → 마지막 글자
- `s[-2]` → 뒤에서 두 번째

아래를 직접 실행해 보세요.


````python
s = "Python"
print("s[0] =", s[0])
print("s[1] =", s[1])
print("s[-1] =", s[-1])
print("s[-2] =", s[-2])
````

### 2-1) 슬라이싱: 부분을 잘라내기

슬라이싱 기본 형태:
`문자열[start : end]`

- start 포함
- end는 포함하지 않음 (여기가 가장 헷갈립니다)

예: `s = "Python"`  
- `s[0:2]` → `"Py"` (0,1까지만)
- `s[2:6]` → `"thon"`


````python
s = "Python"
print(s[0:2])
print(s[2:6])
print(s[:2])   # start 생략 = 처음부터
print(s[2:])   # end 생략 = 끝까지
````

### 2-2) step(간격)까지 포함한 슬라이싱

형태:
`문자열[start : end : step]`

- step=2면 한 글자씩 건너뜁니다.
- step=-1이면 뒤집습니다.

이번 수업 목표에 나온 패턴을 직접 써 보겠습니다.


#### ✅ 자주 쓰는 패턴 예제 (8개)

````python
s = "abcdefg"
print("[::-1] 뒤집기:", s[::-1])
````

````python
s = "abcdefg"
print("[1::2] 1번부터 2칸씩:", s[1::2])
````

````python
s = "abcdefg"
print("[-5:] 뒤에서 5글자:", s[-5:])
````

````python
s = "0123456789"
print("[::2] 짝수 인덱스:", s[::2])
````

````python
s = "0123456789"
print("[1::2] 홀수 인덱스:", s[1::2])
````

````python
s = "Python"
print("[::-2] 뒤에서 2칸씩:", s[::-2])
````

````python
s = "Hello, World!"
print("[7:12] 부분 추출:", s[7:12])
````

````python
s = "Korea"
print("[0:100] end가 넘어가도 괜찮음:", s[0:100])
````

### 2-3) 헷갈리기 쉬운 상황(매우 중요)

#### (1) end는 포함하지 않습니다
`s[0:2]`는 0과 1까지만입니다. 2는 포함되지 않습니다.

#### (2) 인덱스가 범위를 넘어가면?
- 인덱싱 `s[999]`는 에러입니다(범위 밖).
- 슬라이싱 `s[0:999]`는 에러가 아니라 가능한 만큼만 가져옵니다.

직접 확인해 보세요.


````python
s = "abc"
# print(s[10])  # 주석을 풀면 IndexError
print(s[0:10])  # 슬라이싱은 안전하게 동작
````

## 4) `find()` vs `index()` (중요)

둘 다 “부분 문자열의 위치”를 찾는 기능이지만, 실패했을 때 행동이 다릅니다.

- `find()` : 없으면 **-1**을 반환 (안전하게 분기 가능)
- `index()` : 없으면 **에러(ValueError)** 발생

언제 어떤 걸 쓰면 좋을까요?
- “없을 수도 있다” → `find()`가 편합니다.
- “반드시 있어야 한다(없으면 바로 문제)” → `index()`로 강하게 체크할 수 있습니다.


#### ✅ `find()`/`index()` 비교 예제 (8개)

````python
s = "Hello Python"
print("find:", s.find("Python"))
print("find(없음):", s.find("Java"))
````

````python
s = "Hello Python"
print("index:", s.index("Python"))
````

````python
# index는 없으면 에러가 납니다(주석을 풀면 ValueError)
# s = "Hello Python"
# print(s.index("Java"))
````

````python
# find는 없을 때 -1이므로 if로 처리 가능
s = "Hello Python"
pos = s.find("Java")
if pos == -1:
    print("없습니다!")
else:
    print("위치:", pos)
````

````python
# index를 안전하게 쓰려면 try/except를 씁니다.
s = "Hello Python"
try:
    pos = s.index("Java")
    print("위치:", pos)
except ValueError:
    print("index: 찾지 못했습니다!")
````

````python
# 실전 예시: 이메일에서 @ 위치 찾기
email = "abc@gmail.com"
at = email.find("@")
print("at 위치:", at)
print("아이디:", email[:at])
print("도메인:", email[at+1:])
````

````python
# (헷갈림) find는 0도 가능한 값입니다!
# 찾은 위치가 0이면 '찾았다'입니다. -1만 '못 찾았다'입니다.
s = "abc"
print(s.find("a"))  # 0
print(s.find("z"))  # -1
````

````python
# 그래서 if pos: 같은 식으로 쓰면 안 됩니다(0이 False처럼 취급될 수 있음)
s = "abc"
pos = s.find("a")  # 0
if pos:
    print("찾음(잘못된 판단 가능)")
else:
    print("pos가 0이라서 else로 갑니다(실제로는 찾았는데!)")

# 올바른 판단:
if pos != -1:
    print("올바르게: 찾았습니다!")
````

## 5) Raw 문자열: `r'...'` (경로/정규표현식에서 특히 중요)

문자열에는 “이스케이프 문자(escape)”가 있습니다.
예:
- `\n` : 줄바꿈
- `\t` : 탭
- `\\` : 역슬래시 자체를 의미

문제는 “경로”에서 역슬래시 `\`를 많이 쓴다는 점입니다(특히 Windows).

예를 들어, 아래 문자열은 보기엔 경로 같지만 실제로는:
- `\n`이 줄바꿈으로 해석되어 문제가 생길 수 있습니다.

그래서 Raw 문자열 `r'...'`를 쓰면:
- 역슬래시를 특별하게 해석하지 않고 “그대로” 취급합니다.


#### ✅ Raw 문자열 예제 (8개)

````python
# 일반 문자열: \n이 줄바꿈으로 해석될 수 있습니다.
path = "C:\new\test"
print(path)
````

````python
# Raw 문자열: 역슬래시를 그대로 취급합니다.
path = r"C:\new\test"
print(path)
````

````python
# 비교를 위해 repr로 확인해 보기
print(repr("C:\new\test"))
print(repr(r"C:\new\test"))
````

## (정답 예시) — 스스로 푼 뒤에만 보세요

정답은 여러 방식이 있을 수 있습니다. 아래는 “가능한 예시”입니다.


````python
# 불변 + 재할당 예시
s = "python"
s = s[0].upper() + s[1:]
print(s)  # Python

# 슬라이싱 예시
t = "abcdef"
print(t[1:4])  # bcd
print(t[::-1]) # fedcba

# 전처리 예시
line = "  10, 20, 30  "
nums = [int(x.strip()) for x in line.strip().split(",")]
print(nums, sum(nums))

# find/index 예시
email = "abc@gmail.com"
at = email.find("@")
if at != -1:
    print(email[:at], email[at+1:])

# raw 문자열 예시
path = r"C:\new\test"
print(path)
````

````python
# 1. 문장 입력 받기 (문자열)
text = input("문장을 입력하세요: ")

# 2. 양쪽 공백 제거
text = text.strip()

# 3. 모두 소문자로 변환
text = text.lower()

# 4. 쉼표를 공백으로 변환
text = text.replace(",", " ")

# 5. 단어 리스트 만들기 (공백 기준 분리)
words = text.split()

# 6. 단어 개수 출력
print(f"단어 개수: {len(words)}")

# 7. 특정 단어 포함 여부 확인 (대소문자 무시)
if "python" in words:
    print("'python'이 포함되어 있습니다.")
else:
    print("'python'이 포함되어 있지 않습니다.")
````
