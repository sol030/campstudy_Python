    ## 문제 1
    
    **온라인 강의 수강생 관리 시스템**
    
    다음 구조의 데이터를 사용한다.
    
    ```python
    students = {
        "민수": [80, 70, 90],
        "영희": [40, 55, 60],
        "지수": [100, 95, 90],
        "철수": [30, 45, 20]
    }
    
    ```
    
    아래 기능을 함수로 나누어 구현하세요.
    
    필수 함수:
    
    1. 한 학생의 평균 점수를 계산하는 함수
    2. 모든 학생의 평균을 딕셔너리로 만드는 함수
    3. 평균이 60 이상인 학생만 새 딕셔너리로 반환하는 함수
    4. 평균이 가장 높은 학생의 이름을 반환하는 함수
    
    출력 예시:
    
    ```
    학생별 평균: {'민수': 80.0, '영희': 51.6, '지수': 95.0, '철수': 31.6}
    합격자: {'민수': 80.0, '지수': 95.0}
    최고 평균 학생: 지수
    
    ```
    
    조건:
    
    - 원본 students는 절대 변경되면 안 됨
    - 최소 20줄 이상
    
    ---
    
    ### 해설
    
    이 문제는 **딕셔너리 → 리스트 → 평균 계산 → 필터링 → 최대값 찾기**를 함수로 나누는 문제입니다.
    
    원본 students를 직접 수정하지 말고, 계산 결과만 새로운 딕셔너리로 만들어야 합니다.
    
    ---
    
    ### 풀이 1 (기본 반복문 방식)
    
    ```python
    students = {
        "민수": [80, 70, 90],
        "영희": [40, 55, 60],
        "지수": [100, 95, 90],
        "철수": [30, 45, 20]
    }
    
    def calc_avg(scores):
        return sum(scores) / len(scores)
    
    def make_avg_dict(students):
        result = {}
        for name in students:
            result[name] = calc_avg(students[name])
        return result
    
    def get_pass(avg_dict):
        result = {}
        for name in avg_dict:
            if avg_dict[name] >= 60:
                result[name] = avg_dict[name]
        return result
    
    def get_top_student(avg_dict):
        max_name = ""
        max_score = 0
        for name in avg_dict:
            if avg_dict[name] > max_score:
                max_score = avg_dict[name]
                max_name = name
        return max_name
    
    avg_dict = make_avg_dict(students)
    print("학생별 평균:", avg_dict)
    print("합격자:", get_pass(avg_dict))
    print("최고 평균 학생:", get_top_student(avg_dict))
    
    ```
    
    ---
    
    ### 풀이 2 (내장함수 활용)
    
    ```python
    def get_top_student(avg_dict):
        return max(avg_dict, key=avg_dict.get)
    
    ```
    
    이 방식은 딕셔너리에서 값이 가장 큰 키를 자동으로 찾아줍니다.

    ## 문제 2
    
    **은행 계좌 시뮬레이터**
    
    다음 구조를 사용한다.
    
    ```python
    accounts = {
        "A001": 50000,
        "A002": 120000,
        "A003": 30000
    }
    
    ```
    
    다음 기능을 함수로 구현하세요.
    
    필수 기능:
    
    1. 입금 함수
    2. 출금 함수 (잔액 부족 시 출금 실패 처리)
    3. 전체 잔액 합계 계산 함수
    4. 잔액이 50000 미만인 계좌만 모아 반환하는 함수
    
    출력 예시:
    
    ```
    A001 잔액: 70000
    A003 출금 실패
    전체 은행 자산: 220000
    위험 계좌: ['A003']
    
    ```
    
    조건:
    
    - 함수 안에서 딕셔너리를 직접 수정해야 함
    - 매개변수로 딕셔너리를 전달해서 처리
    - 20줄 이상
    
    ---
    
    ### 해설
    
    이 문제는 **딕셔너리를 함수 안에서 직접 변경하는 참조 구조**를 이해해야 합니다.
    
    ---
    
    ### 풀이
    
    ```python
    def deposit(accounts, acc, money):
        accounts[acc] += money
    
    def withdraw(accounts, acc, money):
        if accounts[acc] < money:
            print(acc, "출금 실패")
            return
        accounts[acc] -= money
    
    def total_money(accounts):
        return sum(accounts.values())
    
    def risky_accounts(accounts):
        result = []
        for acc in accounts:
            if accounts[acc] < 50000:
                result.append(acc)
        return result
    
    accounts = {
        "A001": 50000,
        "A002": 120000,
        "A003": 30000
    }
    
    deposit(accounts, "A001", 20000)
    print("A001 잔액:", accounts["A001"])
    
    withdraw(accounts, "A003", 50000)
    
    print("전체 은행 자산:", total_money(accounts))
    print("위험 계좌:", risky_accounts(accounts))
    
    ```

    ## 문제 3
    
    **텍스트 로그 분석기**
    
    아래 문자열이 로그 데이터다.
    
    ```python
    log = "error login success login error logout success login error"
    
    ```
    
    아래 기능을 함수로 구현하세요.
    
    필수 기능:
    
    1. 단어별 횟수를 딕셔너리로 만드는 함수
    2. 가장 많이 나온 단어를 반환하는 함수
    3. 특정 단어의 비율(%)을 계산하는 함수
    4. 횟수가 2회 이상인 단어만 새 딕셔너리로 반환하는 함수
    
    출력 예시:
    
    ```
    단어 빈도: {'error': 3, 'login': 3, 'success': 2, 'logout': 1}
    최다 단어: error
    error 비율: 33.3%
    중요 단어: {'error': 3, 'login': 3, 'success': 2}
    
    ```
    
    조건:
    
    - 문자열 → 리스트 → 딕셔너리 흐름을 함수로 분리
    - 최소 20줄 이상
    
    ---
    
    ### 해설
    
    이 문제는 **문자열 split → 리스트 순회 → 딕셔너리 누적 → 통계 계산** 구조입니다.
    
    ---
    
    ### 풀이
    
    ```python
    def make_list(log):
        return log.split()
    
    def make_count(words):
        result = {}
        for w in words:
            if w in result:
                result[w] += 1
            else:
                result[w] = 1
        return result
    
    def most_word(counts):
        return max(counts, key=counts.get)
    
    def ratio(counts, word):
        total = sum(counts.values())
        return counts[word] / total * 100
    
    def important(counts):
        result = {}
        for k in counts:
            if counts[k] >= 2:
                result[k] = counts[k]
        return result
    
    log = "error login success login error logout success login error"
    
    words = make_list(log)
    counts = make_count(words)
    
    print("단어 빈도:", counts)
    print("최다 단어:", most_word(counts))
    print("error 비율:", round(ratio(counts, "error"), 1), "%")
    print("중요 단어:", important(counts))
    
    ```
    
    ## 문제 4
    
    **1. 실제 출력값**
    
    ```python
    []
    
    ```
    
    ---
    
    **2. 왜 이렇게 나오는가?**
    
    이 코드는 에러 없이 실행되지만, `scores` 안에 아무 값도 들어가지 않습니다.
    
    그 이유는 `add_score` 함수 안에서 일어나는 일이 **원본 리스트를 수정하는 것이 아니라, 새로운 리스트를 만들어서 지역 변수에 대입만 하기 때문**입니다.
    
    ```python
    def add_score(data, x):
        data = data + [x]
    ```
    
    이 한 줄에서 일어나는 일을 단계별로 보면 다음과 같습니다.
    
    1. `data + [x]` 는
        
        기존 리스트 `data`에 `[x]`를 붙인 **새로운 리스트**를 생성합니다.
        
    2. 그 새로운 리스트를 **지역 변수 `data`에 다시 저장**합니다.
    3. 하지만 이 `data`는 함수 안에서만 존재하는 **지역 변수**입니다.
    4. 함수가 끝나면 이 `data`는 사라집니다.
    5. 함수 밖에 있는 `scores`는 아무 변화도 겪지 않습니다.
    
    즉,
    
    ```python
    data = data + [x]
    
    ```
    
    는
    
    "리스트를 수정"하는 것이 아니라
    
    "새 리스트를 만들어서 지역 변수 data가 가리키게 바꾸는 것"입니다.
    
    **3. 함수 호출 흐름**
    
    ```python
    scores = []
    process() 실행
    ```
    
    ```python
    for i in range(3):
        add_score(scores, i * 10)
    ```
    
    각 호출을 보면:
    
    | i | 전달되는 x | 함수 안에서 일어나는 일 |
    | --- | --- | --- |
    | 0 | 0 | data = [] + [0] → data는 [0] (scores는 여전히 []) |
    | 1 | 10 | data = [] + [10] → data는 [10] |
    | 2 | 20 | data = [] + [20] → data는 [20] |
    
    하지만 이 모든 `data`는 **지역 변수**이므로
    
    `scores`는 처음 만든 `[]` 그대로 유지됩니다.
    
    그래서 출력은
    
    ```python
    []
    ```
    
    이 됩니다.

    ## 문제 5
    
    **1. 실제 출력값**
    
    ```python
    [0, 10, 20]
    
    ```
    
    **2. 왜 문제 4와 결과가 다른가?**
    
    이번에는 `add_score`가 이렇게 생겼습니다.
    
    ```python
    def add_score(data, x):
        data.append(x)
    
    ```
    
    여기서 중요한 차이는:
    
    | 문제 4 | 문제 5 |
    | --- | --- |
    | `data = data + [x]` | `data.append(x)` |
    | 새 리스트 생성 | 기존 리스트를 직접 수정 |
    | 지역 변수만 변경 | 원본 객체가 변경 |
    
    `append()`는 **새 리스트를 만들지 않습니다.**
    
    `scores`가 가리키는 리스트 객체 자체를 직접 바꿉니다.
    
    ---
    
    **3. 실행 흐름**
    
    초기 상태:
    
    ```python
    scores = []
    
    ```
    
    반복:
    
    | i | x | append 결과 |
    | --- | --- | --- |
    | 0 | 0 | [] → [0] |
    | 1 | 10 | [0] → [0, 10] |
    | 2 | 20 | [0, 10] → [0, 10, 20] |
    
    `scores`가 가리키는 같은 리스트 객체가 계속 수정되므로
    
    최종 출력은
    
    ```python
    [0, 10, 20]
    
    ```
    
    이 됩니다.
    
    ---
    
    ## 문제 6
    
    **1. 실제 출력값**
    
    ```python
    [0, 10, 20]
    
    ```
    
    ---
    
    **2. 왜 이 코드는 위와 전혀 다른 방식으로 동작하는가?**
    
    이 코드에서는 리스트를 직접 수정하지 않고,
    
    **매번 새 리스트를 만들어서 scores에 다시 저장**합니다.
    
    ```python
    def add_score(data, x):
        return data + [x]
    
    ```
    
    이 함수는:
    
    - 기존 리스트를 바꾸지 않고
    - 새로운 리스트를 만들어
    - 그 새 리스트를 반환합니다.
    
    ---
    
    **3. 핵심: global scores**
    
    ```python
    def process():
        global scores
        for i in range(3):
            scores = add_score(scores, i * 10)
    
    ```
    
    여기서 `scores = ...` 는
    
    **전역 변수 `scores` 자체를 새 리스트로 교체**합니다.
    
    반복 흐름:
    
    | i | 기존 scores | 반환값 | 새 scores |
    | --- | --- | --- | --- |
    | 0 | [] | [0] | [0] |
    | 1 | [0] | [0, 10] | [0, 10] |
    | 2 | [0, 10] | [0, 10, 20] | [0, 10, 20] |
    
    즉,
    
    - 문제 4: 새 리스트를 만들어도 **버림**
    - 문제 5: 같은 리스트를 **직접 수정**
    - 문제 6: 새 리스트를 만들어서 **전역 변수에 다시 저장**
    
    이 차이 때문에 결과가 완전히 달라집니다.
