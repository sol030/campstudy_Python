### 문제 10

`random` 모듈을 사용하여 **주사위 프로그램**을 만들어 보세요.

요구사항:

- 1부터 6 사이의 숫자를 하나 랜덤으로 뽑는 함수를 만든다.
- 뽑힌 숫자를 화면에 출력한다.
- 이 함수를 3번 호출해서 주사위를 3번 굴린 결과를 출력한다.

출력 예시:

```
주사위 결과: 4
주사위 결과: 1
주사위 결과: 6
```

- `random` 모듈
    
    ## random 모듈이란?
    
    `random`은 **무작위(랜덤)** 값을 만들어 주는 파이썬 내장 모듈이다.
    
    게임, 추첨, 시뮬레이션, 테스트 데이터 만들 때 사용한다.
    
    사용하려면 반드시 먼저 불러와야 한다.
    
    ```python
    import random
    ```
    
    ---
    
    ## 가장 많이 쓰는 기능
    
    ### 1) `random.randint(a, b)`
    
    a부터 b까지 **정수 하나를 랜덤으로 반환**
    
    ```python
    import random
    
    n = random.randint(1, 10)
    print(n)   # 1 ~ 10 중 아무 숫자
    
    ```
    
    주의: 끝 숫자 b도 포함된다.
    
    ---
    
    ### 2) `random.random()`
    
    0 이상 1 미만의 **실수 하나를 랜덤으로 반환**
    
    ```python
    import random
    
    x = random.random()
    print(x)   # 예: 0.3849201
    
    ```
    
    ---
    
    ### 3) `random.choice(리스트)`
    
    리스트 안에서 **하나를 랜덤으로 뽑아줌**
    
    ```python
    import random
    
    fruits = ["apple", "banana", "orange"]
    print(random.choice(fruits))
    
    ```
    
    ---
    
    - 그 이외 함수들
        
        
        ### 4) `random.shuffle(리스트)`
        
        리스트의 순서를 **랜덤으로 섞음**
        
        ```python
        import random
        
        nums = [1, 2, 3, 4, 5]
        random.shuffle(nums)
        print(nums)
        
        ```
        
        주의: 새 리스트를 만드는 게 아니라, **원본 리스트를 직접 바꾼다.**
        
        ---
        
        ### 5) `random.sample(리스트, 개수)`
        
        리스트에서 **중복 없이 여러 개를 뽑음**
        
        ```python
        import random
        
        nums = [1, 2, 3, 4, 5]
        picked = random.sample(nums, 3)
        print(picked)
        
        ```
        
    
    ## random 모듈 사용 패턴 요약
    
    ```python
    import random
    
    random.randint(1, 10)      # 숫자 뽑기
    random.choice(["a","b"])  # 목록에서 하나 뽑기
    random.shuffle(list)      # 순서 섞기
    random.sample(list, 3)    # 여러 개 뽑기
    ```
    
    ---
    
    ## 초보자가 제일 많이 틀리는 것
    
    ### 틀린 코드
    
    ```python
    randint(1, 10)   # 오류 발생
    ```
    
    ### 올바른 코드
    
    ```python
    import random
    random.randint(1, 10)
    ```
    
    모듈을 불러왔으면 **random.** 을 꼭 붙여야 한다.
